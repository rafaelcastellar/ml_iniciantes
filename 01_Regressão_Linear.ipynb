# Regress√£o Linear

Quando vemos uma rela√ß√£o em um diagrama de dispers√£o, podemos usar uma reta para resumir essa rela√ß√£o nos dados. Tamb√©m podemos usar essa reta para fazer previs√µes a partir dos dados. Este processo √© chamado de **regress√£o linear**.

## Como ajustar uma reta aos dados

## Uso de equa√ß√µes para retas de regress√£o
Depois de ajustarmos uma reta aos dados, encontramos sua equa√ß√£o e a usamos para fazer previs√µes.

# *y = b + m.x*

>**y** = valor do eixo y de determinado ponto da reta. √â a nossa vari√°vel dependente.
> 
>**b** = O coeficiente linear n da reta r √© definido como o ponto onde a reta intercepta o eixo y, ou seja o ponto de coordenadas P(0,n).
>
>**m** = Podemos encontrar uma equa√ß√£o da reta r conhecendo a sua inclina√ß√£o (dire√ß√£o), ou seja o valor do √¢ngulo Œ∏ que a reta apresenta em rela√ß√£o ao eixo x.
>
>**x** = valor do eixo x de dertminado ponto da reta. No nosso caso, √© a var√≠avel de entrada

### **3. Modelo Regress√£o Linear M√∫ltipla (MLR)**

Separando var√≠aveis independentes (X) da independente (y)

```python
X1_var = df[['Engine Size(L)','Fuel Consumption Comb (L/100 km)','Fuel Consumption Hwy (L/100 km)','Fuel Consumption City (L/100 km)']]
y_var = df['CO2 Emissions(g/km)'] # dependent variable
```

Importando scikit-learn, separando dados de treinamento e teste

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X1_var, y_var, test_size = 0.3, random_state = 0)
```

Treinando modelo

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

Realizando a predi√ß√£o

```python
yhat = lr.predict(X_test)

yhat
```

Para verificar a precis√£o do modelo, podemos calcular a pontua√ß√£o R-quadrada e podemos introduzir uma nova maneira que √© por gr√°fico de distribui√ß√£o.

```python
print('R^2 Treino: ', lr.score(X_train, y_train))
print('R^2 Teste: ', lr.score(X_test, y_test))
```

A seguir vemos o interecpto, que √© o ponto de intercepta√ß√£o da reta no eixo Y, ou seja, quando x=0. √â o *b0* da da nossa equa√ß√£o de regress√£o linear m√∫ltipla.


```python
print('Intercepto:', lr.intercept_)
```

 Vejamos como cada vari√°vel independente afeta a vari√°vel dependente:

```python
coeff_df = pd.DataFrame(lr.coef_,X1_var.columns,columns=['Coefficient'])
coeff_df
```

O segundo m√©todo para verificar a precis√£o do modelo MLR scikit-learn √© construindo um gr√°fico de distribui√ß√£o combinando os valores previstos e os valores reais.

```python
plt.figure(figsize=(10,4))
sb.kdeplot(yhat,  color = 'r', label = 'Predicted Values')
sb.kdeplot(y_test, color = 'b', label = 'Actual Values')
plt.title('Actual vs Predicted Values')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend(loc = 'upper left',)

plt.savefig('ap.png')
```

Este gr√°fico de distribui√ß√£o revela que nossos valores de previs√£o foram executados quase que precisamente em rela√ß√£o aos nossos valores reais, mas h√° alguns outliers que podem ser observados. Isso ocorre porque constru√≠mos um modelo muito b√°sico de regress√£o linear para prever com precis√£o os resultados.

### **4. Fazendo Predi√ß√µes**
